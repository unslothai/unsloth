```python

#@markdown ### 🔑 How to get your ngrok authtoken
#@markdown 1. Go to [ngrok.com](https://ngrok.com) and sign up (free).
#@markdown 2. Once logged in, go to [Your Authtoken](https://dashboard.ngrok.com/get-started/your-authtoken).
#@markdown 3. Copy the token shown on that page.
#@markdown 4. Paste it in the **ngrok_authtoken** field below (only needed for pyngrok).
#@markdown
#@markdown ### ⚖️ Difference between iframe and pyngrok
#@markdown - **iframe**
#@markdown   • Shows the app directly inside this Colab notebook.
#@markdown   • No external URLs – only you can see it.
#@markdown   • Limited by iframe height; some apps may not render perfectly.
#@markdown   • No extra setup needed.
#@markdown - **pyngrok**
#@markdown   • Creates a public, shareable URL (e.g., `https://xxxx.ngrok.io`).
#@markdown   • Accessible from any browser, even outside Colab.
#@markdown   • No iframe constraints – the app looks exactly as it would on a server.
#@markdown   • Requires a free ngrok account and authtoken.
#@markdown   • The URL changes each time you run the cell (free plan).
#@markdown ---
method = "pyngrok"  #@param ["pyngrok", "iframe"]
ngrok_authtoken = ""  #@param {type:"string"}

import time

if method == "pyngrok":
    if not ngrok_authtoken.strip():
        raise ValueError("Please paste your ngrok authtoken in the field above. "
                         "Get it from https://dashboard.ngrok.com/get-started/your-authtoken")

    # Install pyngrok if needed
    try:
        import pyngrok
    except ImportError:
        !pip install -q pyngrok
        import pyngrok

    from pyngrok import ngrok

    # Set the authtoken (one-time per session)
    ngrok.set_auth_token(ngrok_authtoken.strip())

    # Start tunnel
    public_url = ngrok.connect(8888).public_url
    print(f"🚀 Public URL: {public_url}")

    # Keep alive
    for _ in range(10000):
        time.sleep(300)
        print("=", end="")

else:
    # iframe method
    from google.colab import output
    output.serve_kernel_port_as_iframe(8888, height=1200, width="100%")
    for _ in range(10000):
        time.sleep(300)
        print("=", end="")
```
